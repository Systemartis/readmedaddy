# Tidal microplastic flux: reproducing the headline figure

This notebook reproduces Figure 2 of the accompanying paper: estimated
microplastic flux across a tidal estuary as a function of tide phase. Run
top-to-bottom against the dataset described in `data/DATASET.md`.

```python
import csv, statistics

with open('data/samples.csv') as f:
    rows = list(csv.DictReader(f))
print(f'{len(rows)} samples loaded')
```

```python
# Mean particle count per litre, grouped by tide phase.
phases = {}
for r in rows:
    phases.setdefault(r['tide_phase'], []).append(float(r['particles_per_l']))
for phase, vals in sorted(phases.items()):
    print(phase, round(statistics.mean(vals), 2))
```

**Result.** Flux peaks on the ebb tide, consistent with the paper's reported
2.3x enrichment over slack water (see `results/figure-2-note.md`).
